```python
#printing our responses
print(response.headers)
```

```python
#creating dictionaries of the responses
data_dict = response.json()
data_dict
```

```python
# convert to dataframe
data = []
for business in response.json()["businesses"]:
    name = business["name"]
    rating = business["rating"]
    address = business["location"]["display_address"][0]
    zip_code = business["location"]["zip_code"]
    data.append([name, rating, address, zip_code])
```

```python
#dataframe with specific column headers
df = pd.DataFrame(data, columns=["Name", "Rating", "Address", "Zip Code"])
```
